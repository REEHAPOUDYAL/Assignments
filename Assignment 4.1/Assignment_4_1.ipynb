```python
import google.generativeai as genai
from google.colab import userdata

GOOGLE_API_KEY = userdata.get('GOOGLE_API_KEY')
genai.configure(api_key=GOOGLE_API_KEY)

def call_llm(prompt: str, model="gemini-1.5-flash", temperature=0.7):
    print(f"Attempting to call model: '{model}' with prompt (first 50 chars): '{prompt[:50]}...'")
    try:
        model_instance = genai.GenerativeModel(model)
        response = model_instance.generate_content(
            prompt,
            generation_config=genai.types.GenerationConfig(
                temperature=temperature,
            ),
            request_options={"timeout": 60}
        )
        return response.text.strip()
    except Exception as e:
        print(f"\nERROR: Failed to call LLM with model '{model}' for prompt: '{prompt[:100]}...'")
        print(f"Full Error Details: {e}")
        if hasattr(e, 'response') and e.response:
            print(f"Response status code: {e.response.status_code}")
            print(f"Response text: {e.response.text}")
        return f"ERROR: {e}"

prompts = {
    "Summary": "Please summarize the following short paragraph about the benefits of remote work: Remote work offers numerous advantages including increased flexibility, reduced commute times, and potential for better work-life balance. Companies can also benefit from a wider talent pool and lower office overheads.",
    "Analysis": "Analyze the potential downsides of renewable energy sources mentioned in the following text: While solar power is abundant, its intermittency means it's not always available, requiring significant battery storage solutions. Wind farms can face public opposition due to visual impact and noise, and their efficiency varies greatly with weather conditions."
}

results = {label: call_llm(text) for label, text in prompts.items()}
title = "Renewable Energy and Remote Work"
source = "AI Insights Blog"
print("="*80)
print(f"Article selected — {title} [{source}]\n")
for label, summary in results.items():
    print(f"--- {label} Prompt Output ---\n{summary}\n")
```

```python
import google.generativeai as genai
from google.colab import userdata

GOOGLE_API_KEY = userdata.get('GOOGLE_API_KEY')
genai.configure(api_key=GOOGLE_API_KEY)

def call_llm(prompt: str, model="gemini-1.5-flash", temperature=0.2):
    print(f"\n--- Calling Model with Prompt (first 100 chars): '{prompt[:100]}...' ---")
    try:
        model_instance = genai.GenerativeModel(model)
        response = model_instance.generate_content(
            prompt,
            generation_config=genai.types.GenerationConfig(
                temperature=temperature,
            ),
            request_options={"timeout": 60}
        )
        return response.text.strip()
    except Exception as e:
        print(f"\nERROR: Failed to call LLM with model '{model}' for prompt: '{prompt[:100]}...'")
        print(f"Full Error Details: {e}")
        if hasattr(e, 'response') and e.response:
            print(f"Response status code: {e.response.status_code}")
            print(f"Response text: {e.response.text}")
        return f"ERROR: {e}"

customer_review = "This product arrived quickly and was easy to set up. However, the battery life is surprisingly short, making it inconvenient for daily use. I had higher expectations."
direct_prompt = f"""Analyze the sentiment of the following customer review. Respond with only 'Positive', 'Negative', or 'Neutral'.

Review: "{customer_review}"
Sentiment:"""

few_shot_prompt = f"""Analyze the sentiment of the following customer reviews. Respond with only 'Positive', 'Negative', or 'Neutral'.

Review: "I absolutely love this new phone! The camera is amazing and the screen is beautiful."
Sentiment: Positive

Review: "The delivery was slow and the item arrived damaged. Very disappointed."
Sentiment: Negative

Review: "The service was okay, nothing special, but it got the job done."
Sentiment: Neutral

Review: "{customer_review}"
Sentiment:"""

chain_of_thought_prompt = f"""Analyze the sentiment of the following customer review step-by-step. First, identify positive phrases. Second, identify negative phrases. Third, consider if there are any neutral or mixed points. Finally, determine the overall sentiment (Positive, Negative, or Neutral). Show your reasoning, then state the final sentiment on a new line.

Review: "{customer_review}"
"""
print("="*80)
print("Starting Sentiment Analysis with Different Prompting Techniques")
print("="*80)

print("\n--- Direct Prompt Output ---")
direct_output = call_llm(direct_prompt)
print(direct_output)

print("\n--- Few-Shot Prompt Output ---")
few_shot_output = call_llm(few_shot_prompt)
print(few_shot_output)

print("\n--- Chain-of-Thought Prompt Output ---")
chain_of_thought_output = call_llm(chain_of_thought_prompt)
print(chain_of_thought_output)

print("\n" + "="*80)
print("All prompt outputs generated.")
print("Now, you can evaluate these outputs and write your report.")
print("="*80)
```

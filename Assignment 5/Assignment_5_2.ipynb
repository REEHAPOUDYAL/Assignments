```python
from transformers import BlipProcessor, BlipForConditionalGeneration
from PIL import Image
import torch
import gradio as gr
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
processor = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-base")
model = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-base").to(device)
```

```python
def generate_caption(image, prompt="a photography of"):
    inputs = processor(image, prompt, return_tensors="pt").to(device)
    out = model.generate(**inputs)
    caption = processor.decode(out[0], skip_special_tokens=True)
    return caption
```

```python
iface = gr.Interface(
    fn=generate_caption,
    inputs=[
        gr.Image(type="pil", label="Upload Image"),
        gr.Textbox(value="a photography of", label="Prompt"),
    ],
    outputs=gr.Textbox(label="Generated Caption"),
    title="BLIP Image Captioning Demo",
    description="Upload an image and optionally change the prompt to generate a caption using BLIP."
)

iface.launch(debug=True)
```
